# Bar Charts Types Examples

## 1- Vertical Bar Graph

```python
px.bar(
    x= ["Cats","Dogs","Birds"],
    y= [  4   ,  6   ,   20   ],
    labels={'x':'Anaimals Type', 'y':'Count'},
    title="Pet Shop Animal Count "
)
```

```python
a = px.bar(
      x = ["Cats","Dogs","Birds"],
      y = [  4   ,  6   ,   20   ],
    labels={'x':'animal type', 'y': 'animal count'},
    title='Pet shope Count'
      )
a.show()
```

```python
# Vertical Bar Chart ( We Make The Data )
bar = px.bar(
    x = ["Cats","Dogs","Birds"],
    y = [  4   ,  6   ,   20   ],
    title="Shop Pets Count" ,
    labels={'x':'population of Animal','y':'Animal Count'},
)
bar.show()
```

# 2- Horizontal Bar Chart


```python
# Horizontal Bar Chart ( We Make The Data )

Countries_Animals_Count={
    'Source': [ "Egypt" , "USA" , "China" ], 
    'Land'  : [   350   ,  200  ,   395   ],  
    'Air'   : [   250   ,  390  ,   400   ],
    'Water' : [   100   ,  250  ,   399   ]
   }  

bar = px.bar(
    
    data_frame = Countries_Animals_Count,
    
    x='Source',
    
    y=["Land","Air","Water"],
    
    title="Animals Source" ,
    
    orientation='v' # New Parameter 'h' For Horizontal 'v' For Vertical
    
)

bar.show()
```

# 3- Grouped Bar Chart

```python
# Groupby Vertical Bar Chart ( We Make The Data )

Countries_Animals_Count={
    'Source': [ "Egypt" , "USA" , "China" ], 
    'Land'  : [   350   ,  200  ,   395   ],  
    'Air'   : [   250   ,  390  ,   400   ],
    'Water' : [   100   ,  250  ,   399   ]
   }   

bar = px.bar(
    
    data_frame=Countries_Animals_Count,
    
    x='Source',
    
    y=["Land","Air","Water"],
    
    title="Animals Source" ,
    
    orientation='v',
    
    barmode='group' # New Parameter ['stack', 'group', 'overlay', 'relative']
    
)

bar.show()
```

# 4- Stacked Bar Chart

```python
# Stacked Vertical Overlay Bar Chart ( We Make The Data ) 
# Data Over And Down ZERO

temp={
    'Source' : [ "Egypt" , "USA" , "China" ], 
    'January': [   12    ,  -15  ,   -5   ],  
    'May'    : [   28    ,   25  ,   30   ],
    'October': [   20    ,   19  ,   15   ]
   } 

bar = px.bar(
    
    data_frame = temp,
    
    x='Source',
    
    y=["January","May","October"],
    
    title="temperature Over Monthes " ,
    
    orientation='v',
    
    barmode='overlay',
    
#    opacity=0.2  # 0 - 1
    # New Parameter For Color View
    
)

bar.show()
```

## Adding Some Information To Multiple Columns Group By

```python
ds = pd.read_csv("Caste.csv")
ds.head()
```

```python
# Get Just Maharashtra City Data
dss = ds.query('state_name=="Maharashtra"')

# Group By Multiple Columns
dss = ds.groupby(['year','gender'],as_index=False)[
    ['detenues','under_trial','convicts','others']].sum()
```

```python
dss
```

```python
px.bar(data_frame=dss, x="year", y ="under_trial",orientation="v",
    
    color = "gender", # New Parameter For Color View

    text = "gender" #New Parameter For Write Sign on Bars
    
)
```

## Some Color Changes


```python
px.bar(data_frame=dss,x="year",y="under_trial",
       orientation="v",barmode='relative',
    
    text="gender",
    
        
#    color="gender", # Discrete Values "Male", "Female"
    
#    color_discrete_map = {"Male": "black" ,"Female":"pink"},   # map your chosen colors
    
    
    color="under_trial", #continuous Values "Numbers"

    color_continuous_scale= px.colors.sequential.Reds,      # https://plotly.com/python/builtin-colorscales/
    
    range_color=[1,450000]                                   # set your own continuous color scale
    
)
```

## Some Style Changes

```python


px.bar(data_frame=dss, x="year", y="under_trial", 
       orientation="v", text="gender", color="gender",
       color_discrete_map={"Male": "blue" ,"Female":"red"},
    
    width=700,  # Change Width
    
    height=500, # Change Height
       
    template='seaborn',       # 'ggplot2', 'seaborn', 'simple_white', 'plotly',
                                  # 'plotly_white', 'plotly_dark', 'presentation',
                                  # 'xgridoff', 'ygridoff', 'gridon', 'none'
    
)
```

## Making Animation

```python
px.bar(data_frame=dss,
       x="gender",
       y="convicts",
       color="gender",
       
       animation_frame = 'year',     # assign marks to animation frames OVER YEARS
       
       
      range_y=[0,400000],           # set range of x-axis
       
)
```
